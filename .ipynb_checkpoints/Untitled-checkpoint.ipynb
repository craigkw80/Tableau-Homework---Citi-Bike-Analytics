```python
import pandas as pd
```

```python
jan2018 = pd.read_csv("Data/JC-201801-citibike-tripdata.csv")
jan2018.head()
```

```python
july2018 = pd.read_csv("Data/JC-201807-citibike-tripdata.csv")
july2018.head()
```

```python
# calculate age
jan2018['age']=2018-jan2018['birth year']
jan2018.head()
july2018['age']=2018-july2018['birth year']
july2018.head()
```

```python
jan2018["gender"].replace({0: "unknown", 1: "male", 2:"female"}, inplace=True)
jan2018.head()
jan2018['gender'].value_counts()
```

```python
july2018["gender"].replace({0: "unknown", 1: "male", 2:"female"}, inplace=True)
july2018.head()
july2018['gender'].value_counts()
```

```python
jan2018["tripduration_min"] = round(jan2018["tripduration"]/60,2)
jan2018.head()
```

```python
july2018["tripduration_min"] = round(july2018["tripduration"]/60,2)
july2018.head()
```

```python
# combine dataset
combined = pd.concat([jan2018, july2018])
combined.head()
combined.describe() #comfirmed that data concatenated
```

```python
jan2018['ride_id'] = jan2018.index
jan2018.head()

july2018['ride_id'] = july2018.index
july2018.head()

combined['ride_id'] = combined.index
combined.head()
```

```python
jan2018.to_csv("citibikeJan2018.csv")
july2018.to_csv("citibikeJuly2018.csv")
combined.to_csv("combined_citibike.csv")
```
